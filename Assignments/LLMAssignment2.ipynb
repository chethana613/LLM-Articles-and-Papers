***Metrics Explored:***

*   Accuracy - Measures the proportion of correctly classified instances among all instances.

*   Precision - Quantifies the proportion of true positive predictions among all positive predictions made.

*   Recall - Calculates the proportion of true positive predictions among all actual positive instances.

*   F1-score - It's the Harmonic mean of precision and recall, providing a balanced measure of a classifier's performance.

```python
#Zero shot Bloom 560m - Basic prompt

from transformers import AutoModelForSequenceClassification, AutoTokenizer
from datasets import load_dataset
from sklearn.metrics import accuracy_score, classification_report, precision_recall_fscore_support
import torch
from tqdm import tqdm

dataset = load_dataset("cardiffnlp/tweet_sentiment_multilingual", "english", split='test')

model_name = "bigscience/bloom-560m"
#model_name = "bigscience/bloom-1b1"
#model_name = "bigscience/bloom-1b7"
#model_name = "bigscience/bloomz-560m"
#model_name = "bigscience/bloomz-1b1"
#model_name = "bigscience/bloomz-1b7"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

labels = ["negative", "neutral", "positive"]

prompt = 'Classify the given text into one of the following three sentiments: neutral, negative, positive.'

true_labels = []
predicted_labels = []

# Wrapping the loop with tqdm for progress
for text, label_id in tqdm(zip(dataset['text'], dataset['label']), total=len(dataset)):
    true_label = labels[label_id]
    true_labels.append(true_label)

    inputs = tokenizer(prompt + " " + text, return_tensors="pt", padding=True, truncation=True)
    outputs = model(**inputs)
    predicted_label_idx = torch.argmax(outputs.logits, dim=1).item()
    predicted_label = labels[predicted_label_idx]
    predicted_labels.append(predicted_label)

accuracy = accuracy_score(true_labels, predicted_labels)
metrics = precision_recall_fscore_support(true_labels, predicted_labels, average='weighted')
classification_report_output = classification_report(true_labels, predicted_labels, target_names=labels)

print(f"Accuracy: {accuracy}")
print(f"Precision: {metrics[0]}")
print(f"Recall: {metrics[1]}")
print(f"F1-score: {metrics[2]}")
print("Classification Report:")
print(classification_report_output)
```

```python
#Few Shot Bloom 560m k=2

import warnings
warnings.filterwarnings('ignore')
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from datasets import load_dataset
from sklearn.metrics import accuracy_score, classification_report, f1_score, precision_score, recall_score
import torch
from collections import defaultdict
from tqdm import tqdm


model_name = "bigscience/bloom-560m"
#model_name = "bigscience/bloom-1b1"
#model_name = "bigscience/bloom-1b7"
#model_name = "bigscience/bloomz-560m"
#model_name = "bigscience/bloomz-1b1"
#model_name = "bigscience/bloomz-1b7"

train_dataset = load_dataset("cardiffnlp/tweet_sentiment_multilingual", "english", split='train')
test_dataset = load_dataset("cardiffnlp/tweet_sentiment_multilingual", "english", split='test')

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

labels = ["negative", "neutral", "positive"]


prompt = 'Classify the given text into one of the following three sentiments : neutral,negative,positive.'

k = 2
few_shot_predictions = []
true_labels = []


few_shot_examples = defaultdict(list)
for i, (text, label_id) in enumerate(zip(train_dataset['text'], train_dataset['label'])):
    if len(few_shot_examples[labels[label_id]]) < k:
        few_shot_examples[labels[label_id]].append(text)
    if all(len(examples) == k for examples in few_shot_examples.values()):
        break

for text, label_id in tqdm(zip(test_dataset['text'], test_dataset['label']), total=len(test_dataset)):
    true_label = labels[label_id]

    for example_text in few_shot_examples[true_label]:

        inputs = tokenizer(prompt + example_text + text, return_tensors="pt", padding=True, truncation=True)
        outputs = model(**inputs)
        predicted_label_idx = torch.argmax(outputs.logits, dim=1).item()
        predicted_label = labels[predicted_label_idx]
        #print(f"\nPredicted Label:{predicted_label}")
        few_shot_predictions.append(predicted_label)
        true_labels.append(true_label)
        #print(f"\nTrue Label:{predicted_label}")

few_shot_accuracy = accuracy_score(true_labels, few_shot_predictions)
few_shot_f1 = f1_score(true_labels, few_shot_predictions,  average='weighted')
few_shot_precision = precision_score(true_labels, few_shot_predictions, average='weighted')
few_shot_recall = recall_score(true_labels, few_shot_predictions, average='weighted')

few_shot_classification_report = classification_report(true_labels, few_shot_predictions, target_names=labels)

print("\nFew-Shot Results (k=2):")
print(f"Few-Shot Accuracy: {few_shot_accuracy}")
print(f"Few-Shot f1: {few_shot_f1}")
print(f"Few-Shot Precision: {few_shot_precision}")
print(f"Few-Shot Recall: {few_shot_recall}")
print("Few-Shot Classification Report:")
print(few_shot_classification_report)
```

```python
#tried comparing using sentiment analyzer popularity scores

from transformers import AutoModelForCausalLM, AutoTokenizer
from datasets import load_dataset
from tqdm import tqdm
import torch
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

#model_name = "bigscience/bloom-560m"
#model_name = "bigscience/bloom-1b1"
#model_name = "bigscience/bloom-1b7"
model_name = "bigscience/bloomz-560m"
#model_name = "bigscience/bloomz-1b1"
#model_name = "bigscience/bloomz-1b7"
dataset = load_dataset("cardiffnlp/tweet_sentiment_multilingual", "english", split='test')

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
prompt = 'Classify the text into one of the following : neutral,negative,positive.'

analyzer = SentimentIntensityAnalyzer()


generation_params = {
    "num_beams": 5,  # Number of beams for beam search
    "max_new_tokens": 20,  # Maximum number of new tokens generated
    "early_stopping": True,  # Whether to stop generation when all beam hypotheses reached the EOS token
    "temperature": 1.0,  # Softmax temperature for sampling
    "length_penalty": 0.8,  # Length penalty for beam search
    "no_repeat_ngram_size": 3,  # N-gram blocking size to prevent repetition
    "num_return_sequences": 2,  # Number of beam hypotheses to return
    "do_sample": True,  # Whether to use sampling during generation
    "max_length": 10  # Maximum length of the generated sequence
}

id_to_sentiment = {
    0: 'negative',
    1: 'neutral',
    2: 'positive'
}

for text, label_id in tqdm(zip(dataset['text'], dataset['label']), total=len(dataset)):

    inputs = tokenizer.encode(prompt + text, return_tensors="pt")


    outputs = model.generate(inputs, **generation_params)


    generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)


    sentiment_scores = analyzer.polarity_scores(generated_text)

    if sentiment_scores['compound'] >= 0.05:
        predicted_sentiment = 'positive'
    elif sentiment_scores['compound'] <= -0.05:
        predicted_sentiment = 'negative'
    else:
        predicted_sentiment = 'neutral'


    print("Generated Text:", generated_text)
    print("Predicted Sentiment:", predicted_sentiment)
    print("True Sentiment:", id_to_sentiment[label_id])
```
